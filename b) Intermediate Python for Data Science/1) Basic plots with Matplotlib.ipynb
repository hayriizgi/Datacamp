# Basic plots with Matplotlib

It is a Python package for Data Visualization.

import matplotlib.pyplot as plt # imported from sub-package pyplot

```python
import matplotlib.pyplot as plt

year = [1950, 1970, 1990, 2010] # list for year
pop = [2.519, 3.692, 5.263, 6.972] # list for population
```

### Line Chart

When you have a time scale along the horizontal axis, the line plot is a good choice.

```python
plt.plot(year, pop) # plot() functions tells Python what and how to plot - it will produce a line chart
                    # first argument is x-axis and second one is y-axis
plt.show() # Python requires the show() function to display the chart in case may want to add other customizations to the plot
           # suchs as titles, legend, axis names, etc.
```

### Scatter Chart

In many cases, especially when you're trying to assess if there's a correlation between two variables, for example, the scatter plot is the better choice.

**matplotlib.pyplot.scatter
(x, y, s=None, c=None, marker=None, cmap=None, norm=None, vmin=None, vmax=None, alpha=None, linewidths=None, verts=None, edgecolors=None, hold=None, data=None, **kwargs)**
A scatter plot of y vs x with varying marker size and/or color.

*Parameters:*	
- x, y : array_like, shape (n, ) - The data positions.
- s : scalar or array_like, shape (n, ), optional - The marker size in points **2
- c : color, sequence, or sequence of color, optional, default: ‘b’ - The marker color.
- alpha : scalar, optional, default: None - The alpha blending value, between 0 (transparent) and 1 (opaque)

```python
plt.scatter(year, pop) # scatter() functions tells Python to plot a Scatter Chart
                       # first argument is x-axis and second one is y-axis
plt.show()
```

```python
import numpy as np

x = np.random.normal(1, 0.4, 100)
y = np.arange(len(x))

plt.scatter(x,y)
plt.show()
```

```python
plt.xscale('log') # Sets the scaling of the x-axis. xscale(scale, **kwargs) - scale : [‘linear’ | ‘log’ | ‘logit’ | ‘symlog’]
plt.scatter(x,y)
plt.show()
```

```python
plt.scatter(x,y, s=200, c='r', alpha=0.3) # You can use an array for size (s) as well like a third variable like population, etc.)
                                          # You can use an array (list) or dictionary for color (c)
                                          # Alpha sets the opacity
plt.show()
```

**matplotlib.pyplot.text(x, y, s, fontdict=None, withdash=False, **kwargs)*

Add text in string s to axis at location x, y, data coordinates.

*Parameters:*	
- x, y : scalars - data coordinates
- s : string - text
- fontdict : dictionary, optional, default: None - A dictionary to override the default text properties. If fontdict is None, the defaults are determined by your rc parameters.
- withdash : boolean, optional, default: False - Creates a TextWithDash instance instead of a Text instance.

*Other Parameters:*
 	
- kwargs : Text properties. - Other miscellaneous text parameters.

**matplotlib.pyplot.grid(b=None, which='major', axis='both', **kwargs)**

Turn the axes grids on or off. Set the axes grids on or off; b is a boolean

- If b is None and len(kwargs)==0, toggle the grid state. If kwargs are supplied, it is assumed that you want a grid and b is thus set to True.

- which can be ‘major’ (default), ‘minor’, or ‘both’ to control whether major tick grids, minor tick grids, or both are affected.

- axis can be ‘both’ (default), ‘x’, or ‘y’ to control which set of gridlines are drawn.

- kwargs are used to set the grid line properties, e.g. linestyle='-', linewidth=2, etc.

- linestyle or ls - [‘solid’ | ‘dashed’, ‘dashdot’, ‘dotted’ | (offset, on-off-dash-seq) | '-' | '--' | '-.' | ':' | 'None' | ' ' | '']
- linewidth or lw - float value in points

```python
plt.scatter(x,y, s=200, c='r', alpha=0.3) 

plt.text(1, 50, 'India')
plt.grid(True,linestyle='--', linewidth=1)

plt.show()
```

### Histogram

- A histogram is an accurate representation of the distribution of numerical data. It gives you an idea about the distribution of variables.
- It is an estimate of the probability distribution of a continuous variable (quantitative variable)
- To construct a histogram, the first step is to "bin" the range of values—that is, divide the entire range of values into a series of intervals—and then count how many values fall into each interval. The bins are usually specified as consecutive, non-overlapping intervals of a variable. They can be or not be of equal size.
- hist(**x, bins=10**, range=None, normed=False, weights=None, cumulative=False, bottom=None, histtype='bar', align='mid',
 orientation='vertical', rwidth=None, log=False, color=None, label=None, stacked=False, hold=None, data=None, **kwargs)--
 Plots a histogram. 
 - x is the list of values
 - bins is the bin size and by default it is 10

```python
values = [0,0.6,1.4,1.6,2.2,2.5,2.6,3.2,3.5,3.9,4.2,6] 
plt.hist(values, bins=3)
plt.show()
```

### Chart Customizations

```python
plt.plot(year, pop)

plt.xlabel('Year') # add x-axis label
plt.ylabel('Population') # add y-axis label
plt.title('World Population Projections') # adds a Chart Title

plt.show()
```

**xticks(locs, [labels], *kwargs) and yticks(locs, [labels], *kwargs)**

*Parameters:*	
locs : array_like - A list of positions at which ticks should be placed. *You can pass an empty list to disable yticks (yticks([])*

labels : array_like, optional - A list of explicit labels to place at the given locs.

**kwargs - Text properties can be used to control the appearance of the labels.

*Returns:*	
locs - An array of label locations.

labels - A list of Text objects.

```python
plt.plot(year, pop)

plt.xlabel('Year') # add x-axis label
plt.ylabel('Population') # add y-axis label
plt.title('World Population Projections') # adds a Chart Title

plt.yticks([0, 2, 4, 6, 8, 10]) # sets the current tick locations and labels of the y-axis.

plt.show()
```

```python
plt.plot(year, pop)

plt.xlabel('Year') # add x-axis label
plt.ylabel('Population') # add y-axis label
plt.title('World Population Projections') # adds a Chart Title

plt.yticks([0, 2, 4, 6, 8, 10], ['0', '2B', '4B', '6B', '8B', '10B']) # locations and labels for ticks

plt.show()
```

```python
# Add more data

year = [1800, 1850, 1900] + year # add the new data to beginning of the list
pop = [1.0, 1.262, 1.650] + pop # add the new data to beginning of the list

plt.plot(year, pop)

plt.xlabel('Year') # add x-axis label
plt.ylabel('Population') # add y-axis label
plt.title('World Population Projections') # adds a Chart Title

plt.yticks([0, 2, 4, 6, 8, 10], ['0', '2B', '4B', '6B', '8B', '10B']) # locations and lables for ticks

plt.show()
```
